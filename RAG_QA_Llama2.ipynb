## Imports

```python
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
from langchain.chains.summarize import load_summarize_chain
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain import LLMChain
from langchain.prompts.prompt import PromptTemplate
from langchain.llms import HuggingFacePipeline
from langchain.vectorstores import Chroma
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.chains import RetrievalQA
import torch
from transformers import BitsAndBytesConfig
from langchain_community.document_loaders import PyPDFDirectoryLoader
```

## HuggingFace Token

```python
auth_token = "YourHuggingFaceToken"
```

## Model

```python
name = "meta-llama/Llama-2-7b-chat-hf"
```

##Tokenization and Model Initialization

```python
quantization_config = BitsAndBytesConfig(load_in_4bit=True,
                                         bnb_4bit_quant_type="nf4",
                                         bnb_4bit_use_double_quant=True,
                                         bnb_4bit_compute_dtype=torch.bfloat16)
```

```python
tokenizer = AutoTokenizer.from_pretrained(name,cache_dir="./model/", token=auth_token)
```

```python
model = AutoModelForCausalLM.from_pretrained(name,cache_dir="./model/",
                                             device_map = "auto",
                                             torch_dtype=torch.float16,
                                             quantization_config=quantization_config,
                                             token=auth_token)
```

##Pipeline

```python
pipe = pipeline("text-generation",
                model = model,
                tokenizer = tokenizer,
                max_new_tokens=512,
                temperature=0.4,
                top_p=0.95,
                repetition_penalty=1.15)
```

```python
llm = HuggingFacePipeline(pipeline=pipe)
```

## Data Loader

```python
loader = PyPDFDirectoryLoader("data/") #Contains PDFs from https://agriwelfare.gov.in/en/Annual
```

```python
docs = loader.load()
```

##Chunking

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size=100, chunk_overlap=10)
chunks = text_splitter.split_documents(docs)
```

## Embeddings and VectorStore

```python
embedding_model = "sentence-transformers/all-MiniLM-L6-v2"
```

```python
embeddings = HuggingFaceEmbeddings(model_name=embedding_model)
```

```python
vectorsdb = Chroma.from_documents(chunks, embeddings, persist_directory="db")
```

## Create Chain

```python
retriever = vectorsdb.as_retriever()
qa_chain = RetrievalQA.from_chain_type(llm=llm,
                                     chain_type="map_reduce",
                                     retriever=retriever)
```

## Output

```python
def llmqa(chain,query):
    return chain.run(query)
```

```python
print(llmqa(qa_chain,"What is cereal productivity in 2018-19?"))
```

```python
print(llmqa(qa_chain,"How much workforce in India is employed in agricultural activities?"))
```
